```python
import duckdb

# create global connection (optional but recommended)
con = duckdb.connect("saas.db")

def run_sql_checks(path: str):
    """
    Executes all SQL statements inside a file, prints each result.
    """
    # load file
    with open(path, "r") as f:
        sql = f.read()

    # split into individual statements
    queries = [q.strip() for q in sql.split(";") if q.strip()]

    # execute each query
    for i, q in enumerate(queries, start=1):
        print(f"\n------ QUERY {i} ------\n")
        try:
            result = con.execute(q).fetchall()
            print(result)
        except Exception as e:
            print("Error:", e)
```

```python
run_sql_checks("sql/01_check_accounts.sql")
run_sql_checks("sql/02_check_subscriptions.sql")
run_sql_checks("sql/03_check_churn_events.sql")
run_sql_checks("sql/03_check_churn_events.sql")
run_sql_checks("sql/05_check_support_tickets.sql")
```
