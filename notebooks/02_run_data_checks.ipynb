```python
import duckdb

def run_sql_checks(path: str, db_path: str = "saas.db"):
    con = duckdb.connect(db_path)

    with open(path, "r") as f:
        sql = f.read()

    queries = [q.strip() for q in sql.split(";") if q.strip()]

    for i, q in enumerate(queries):
        print(f"\n------ QUERY {i+1} ------\n")
        try:
            # execute separately so we can inspect the result
            rel = con.execute(q)

            # If this query doesn't produce a result set, rel will be None
            # or rel.description will be None.
            if rel is None or getattr(rel, "description", None) is None:
                print("Query executed successfully (no result set).")
            else:
                rows = rel.fetchall()
                print(rows)
        except Exception as e:
            print("Error:", e)

    con.close()
```

```python
run_sql_checks("sql/01_check_accounts.sql")
run_sql_checks("sql/02_check_subscriptions.sql")
run_sql_checks("sql/03_check_churn_events.sql")
run_sql_checks("sql/03_check_churn_events.sql")
run_sql_checks("sql/05_check_support_tickets.sql")
```

```python
import duckdb

con = duckdb.connect("saas.db")

rows = con.execute("""
SELECT table_schema, table_name
FROM information_schema.tables
WHERE table_type='BASE TABLE';
""").fetchall()

if len(rows) == 0:
    print("Database is empty (no tables).")
else:
    print("Tables found:")
    for schema, name in rows:
        print(f"{schema}.{name}")

con.close()
```
